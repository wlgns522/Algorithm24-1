# 선형 조사법

### 1. 문제 정의

 - 문제 이름 : 선형 조사법의 삽입 알고리즘
 - 문제 설명 : 키값을 테이블에 삽입할 때 앞에 있는 키값부터 먼저 맞는 자리에 넣고 같은 자리값이 나온 키는 빈 자리가 나올 때 까지 한칸씩 +하면서 빈 자리를 찾는다
 - 문제 예시 : 테이블이 13일 때 키 45와 71은 똑같이 6번째 자리라서 먼저 입력된 수가 6번 자리에 들어가고 나중에 입력된 수가 7번 자리가 된다.

### 2. 알고리즘 설명

먼저 테이블 크기를 입력받고 None으로 초기화를 한다
그리고 입력된 숫자들을 해시함수에 맞는 주소에 넣는데 중복된 값이 나올 경우 주소를 +1 하면서 빈 자리를 계속 찾는다.
마지막 자리까지 갔는데도 비어있는 곳이 없을 경우 첫번째 자리로 가서 다시 찾는다.

### 4. 알고리즘 코드

```python
M = 13
table = [None]*M
def hashFn(key) :
    return key % M

def lp_insert(key) :
    id = hashFn(key)
    count = M
    while count>0 and (table[id] != None) :
        id = (id +1 +M) & M
        count -=1
    if count > 0 :
        table[id] = key
    return
```

### 5. 테스트 코드

```python
print(" 최초:", table)
lp_insert(45); print("45 삽입:", table)
lp_insert(27); print("27 삽입:", table)
lp_insert(31); print("31 삽입:", table)
lp_insert(12); print("12 삽입:", table)
lp_insert(67); print("67 삽입:", table)
```
