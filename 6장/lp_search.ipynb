# 선형 조사 알고리즘

### 1. 문제 정의

 - 문제 이름 : 선형 조사법의 탐색 알고리즘
 - 문제 설명 : 선형 조사법의 삽입 알고리즘에서 입력한 값을 찾는 알고리즘이다.
 - 문제 예시 : 테이블이 13일 때  키가 45,27,88,9,71,60,46,38,24 일 때 46은 11번 자리에서 탐색에 성공하지만 39는 2번 자리에서 빈 공간이 있기 때문에 실패하게 된다.

### 2. 알고리즘 설명

삽입에서 입력한 값을 탐색에서 다시 입력 하여 저장된 키 값과 입력한 값이 맞을 때 까지 +1 하면서 비교를 한다 찾을 경우 성공하지만 다음 버킷이 비어있을 경우 실패한다.

### 4. 알고리즘 코드

```python
M =13
table = [None]*M
def hashFn(key) :
    return key%M

def lp_insert(key) :
    id = hashFn(key)
    count = M
    while count>0 and (table[id] != None and table[id] != -1) :
        id = (id +1 +M) & M
        count -=1
    if count > 0 :
        table[id] = key
    return

def lp_search(key) :
    id = hashFn(key)
    count = M
    while count>0 :
        if table[id] == None :
            return None
        if table[id] == key :
            return table[id]
        id = (id +1) % M
        count -=1
    return None
```

### 5. 테스트 코드

```python
print(" 최초:", table)
lp_insert(45); print("45 삽입:", table)
lp_insert(27); print("27 삽입:", table)
lp_insert(31); print("31 삽입:", table)
lp_insert(12); print("12 삽입:", table)
lp_insert(67); print("67 삽입:", table)
print("12 탐색:", lp_search(12))
```
